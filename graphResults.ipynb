```python
import matplotlib.pyplot as plt
import numpy as np


def confidence_interval(data):
    #using 95 percent confidence interval
    data_points = len(data)
    average_accuracy = 0
    for i in range(0,data_points):
        average_accuracy = average_accuracy + data[i]/data_points

    standard_deviation = 0
    variance = 0
    for i in range(0,data_points):
        variance = variance + ((data[i] - average_accuracy) * (data[i] - average_accuracy))
    
    standard_deviation = np.sqrt(variance/data_points)
    standard_error = standard_deviation/np.sqrt(data_points)

    return [average_accuracy - 1.96*standard_error, average_accuracy, average_accuracy + 1.96*standard_error]



x = np.arange(2,5)
mixtureOfExperts = np.load("MixtureOfExperts/mixtureOfExpertsResults.npy")

plt.plot(x, mixtureOfExperts[:,0])
plt.plot(x, mixtureOfExperts[:,1])
plt.plot(x, mixtureOfExperts[:,2])

#plt.legend(['lower interval', 'mean', 'higher interval'], loc='upper left')

plt.show()

x = np.arange(2,5)
ensembleMethod = np.load("EnsembleMethod/ensembleMethodResults.npy")

plt.plot(x, ensembleMethod[:,0])
plt.plot(x, ensembleMethod[:,1])
plt.plot(x, ensembleMethod[:,2])

#plt.legend(['lower interval', 'mean', 'higher interval'], loc='upper left')

plt.show()


#'times_to_train.npy'
x = np.arange(2,10)
kmeans_times = np.load("MixtureOfExperts/times_to_train.npy")
interval_for_kmeans = []
for i in kmeans_times:
    inerval_for_kemans.append(confidence_interval(i))

interval_for_kmeans = np.asarray(interval_for_kmeans)

plt.plot(x, interval_for_kmeans[:,0])
plt.plot(x, interval_for_kmeans[:,1])
plt.plot(x, interval_for_kmeans[:,2])

#plt.legend(['lower interval', 'mean', 'higher interval'], loc='upper left')

plt.show()
```
